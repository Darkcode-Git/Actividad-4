```python
def kmeans_analysis(df):
    print("\n4. K-MEANS CLUSTERING ANALYSIS")

    # Prepare data for clustering
    cluster_vars = ['Calorías_quemadas', 'Porcentaje_grasa', 'IMC']
    X = df[cluster_vars]

    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Elbow method
    inertias = []
    K = range(2, 11)

    for k in K:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X_scaled)
        inertias.append(kmeans.inertia_)

    # Plot elbow curve
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(K, inertias, 'bo-')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.title('Elbow Method')
    plt.grid(True)

    # Calculate optimal k
    diffs = np.diff(inertias)
    elbow_point = np.argmin(diffs) + 2
    print(f"\nSuggested optimal number of clusters (elbow method): {elbow_point}")

    # Final clustering
    optimal_kmeans = KMeans(n_clusters=elbow_point, random_state=42)
    clusters = optimal_kmeans.fit_predict(X_scaled)

    # Visualize clusters
    plt.subplot(1, 2, 2)
    scatter = plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis')
    plt.xlabel(cluster_vars[0])
    plt.ylabel(cluster_vars[1])
    plt.title('Cluster Visualization')
    plt.colorbar(scatter)

    plt.tight_layout()
    plt.show()

# Ejecutar análisis de clustering
kmeans_analysis(df)
```
